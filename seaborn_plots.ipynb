```python
import seaborn as sns
%matplotlib inline
```

```python
df=sns.load_dataset('tips')
df
```

```python
sns.distplot(df['total_bill'])
```

```python
sns.jointplot(x='total_bill',y='tip',data=df,kind='scatter')
```

```python
sns.jointplot(x='total_bill',y='tip',data=df,kind='reg')
```

```python
sns.pairplot(df,hue='sex',palette='rainbow')
```

```python
sns.rugplot(df)
```

```python
sns.kdeplot(df)
```

```python
sns.barplot(x='size',y='tip',data=df,hue='sex')
```

```python
sns.boxplot(x='size',y='tip',data=df,hue='sex')
```

```python
sns.violinplot(x='size',y='tip',data=df,hue='sex')
```

```python
sns.swarmplot(x='size',y='tip',data=df,hue='sex')
```

```python
sns.stripplot(x='size',y='tip',data=df,hue='sex')
```
